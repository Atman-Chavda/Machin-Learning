```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Position_Salaries.csv')
x = df.iloc[:,1:-1].values
y = df.iloc[:,-1].values
```

```python
# testing linear modal on given data 
# here we can see that predications are way off 

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x,y)
y_pred = regressor.predict(x)
np.set_printoptions(precision=2)
predected_data = np.concatenate((y_pred.reshape(len(y_pred),1),y.reshape(len(y),1)),1)
print(predected_data)
df1 = pd.DataFrame(predected_data,columns=['predected','real'])
a = df1['real']-df1['predected']
a = np.array(a)
print(a)

# ploting grapg 
plt.scatter(x,y,color = 'red')
plt.plot(x,regressor.predict(x),color = 'blue')
plt.title('Linear Regression')
plt.xlabel('position')
plt.ylabel('salary')
plt.show()
```

```python
df = pd.read_csv('Position_Salaries.csv')
x = df.iloc[:,1:-1].values
y = df.iloc[:,-1].values
```

```python
# applying polynomial modal 


from sklearn.preprocessing import PolynomialFeatures

poly_reg = PolynomialFeatures(degree=2)
X_poly = poly_reg.fit_transform(x)
lin_reg = LinearRegression()
lin_reg.fit(X_poly,y)

plt.scatter(x,y,color = 'red')
plt.plot(x,lin_reg.predict(X_poly),color = 'blue')
plt.title('Linear Regression')
plt.xlabel('position')
plt.ylabel('salary')
plt.show()
```

```python
# predicting the output of single value  
# regressor.predict([[6.5]])
lin_reg.predict(poly_reg.fit_transform([[6.5]]))
```
