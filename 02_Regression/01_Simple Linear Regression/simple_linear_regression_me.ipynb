```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataframe = pd.read_csv('Salary_Data.csv')
x = dataframe.iloc[:,:-1]
# numpy array
y = dataframe.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
print(X_test)
```

```python
# here we're creating simple linear regreation model and test the modal
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# now will try to predict the salary of the test data 
y_pred = regressor.predict(X_test)
# for i in regressor.predict(X_test):
#     print(i)
```

```python
# now we'll plot the graph for better visulisetion
# we'll fo this for predication vs trianing set and prediction vs testset
# regration line is the line which is closest to the real things

plt.scatter(X_train, y_train, color='red') # real salary dots
plt.plot(X_train, regressor.predict(X_train), color='blue') # predicted salary line
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.show()
```

```python
# here is the same thing as above but for test set 
plt.scatter(X_test, y_test, color='red') 
plt.plot(X_train, regressor.predict(X_train),color='blue') # here the line created from this for test set and training set will be the same
# plt.plot(X_test,y_pred,color='green')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.show()
```
