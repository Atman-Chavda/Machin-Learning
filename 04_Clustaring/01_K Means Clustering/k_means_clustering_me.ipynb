```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataframe = pd.read_csv('Mall_Customers.csv')
# here we're takling only last two columns so we can plot 2d graph 
X = dataframe.iloc[:,[3,4]].values
```

Using elbow method to find optimal number of clusters

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    Kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    Kmeans.fit(X)
    wcss.append(Kmeans.inertia_)
plt.plot(range(1,11), wcss)
plt.title('elbow')
plt.xlabel('no of cluster')
plt.ylabel(wcss)
```

```python
Kmeans = KMeans(n_clusters=5,init='k-means++',random_state=42)
y_pred = Kmeans.fit_predict(X)
```

```python
plt.scatter(X[y_pred==0,0], X[y_pred==0,1], s=50, c='red' lable)
plt.scatter(X[y_pred==1,0], X[y_pred==1,1], s=50, c='blue')
plt.scatter(X[y_pred==2,0], X[y_pred==2,1], s=50, c='green')
plt.scatter(X[y_pred==3,0], X[y_pred==3,1], s=50, c='black')
plt.scatter(X[y_pred==4,0], X[y_pred==4,1], s=50, c='yellow')
plt.scatter(Kmeans.cluster_centers_[:,0],Kmeans.cluster_centers_[:,1], s=150, c='magenta')
plt.title('clusters')
plt.xlabel('income')
plt.ylabel('spending score')
```
